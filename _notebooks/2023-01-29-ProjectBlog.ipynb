# My portion
My portion of the project will be sports jeopardy. How it will work is that there will be 5 sports categories and multiple point values that you can answer trivia questions. For example somebody could choose nfl for 300 where they can answer a medium difficulty nfl question for 300 points. If they are correct then they get points if they are wrong somebody else can steal those points from them. The user inputs in their answer which will be then checked if its correct in order for the user to recieve points or not.
