# (0.6 points) Create your own dictionary/list and use sorting algorithms to perform operations on them. (Use playlist maker as an example of what we expect ~ Make something creative and useful)



```python
# Create a dictionary of songs with their lengths
playlist = {
    "Bohemian Rhapsody": 354,
    "Stairway to Heaven": 482,
    "Hotel California": 390,
    "November Rain": 537,
    "Free Bird": 540,
    "In The End": 217,
    "Sweet Child O' Mine": 355,
    "Comfortably Numb": 383,
    "Purple Rain": 508,
    "Let It Be": 243
}

# Sort the songs based on their length (in seconds)
sorted_playlist = sorted(playlist.items(), key=lambda x: x[1])

# Print the sorted playlist
print("Sorted Playlist:")
for song in sorted_playlist:
    print(f"{song[0]} - {song[1]} seconds")
```

# Simulation

```python
import random

# Define the possible outcomes of rolling a six-sided die
dice_outcomes = [1, 2, 3, 4, 5, 6]

# Ask the user how many times they want to roll the dice
num_rolls = int(input("How many times do you want to roll the dice? "))

# Roll the dice the specified number of times and store the results in a list
roll_results = []
for i in range(num_rolls):
    roll = random.choice(dice_outcomes)
    roll_results.append(roll)

# Print the results of the dice rolls
print("Results of rolling the dice:")
for roll in roll_results:
    print(roll)
```
